```python
df_gp_2020 = pd.read_csv("/content/Historical-Spot-Price-GOLD-01-01-2020-to-31-12-2020.csv")
```

```python
df_gp_2020
```

```python
# Keep only required columns
# Clean column names first to remove any leading/trailing whitespace or newline characters
df_gp_2020.columns = df_gp_2020.columns.str.strip()
df_gp_2020.columns = df_gp_2020.columns.str.replace('\n', '')

df_gp_2020 = df_gp_2020[
    ['UPDATED DATE', 'SESSION 1 * SPOT PRICE (IN RS.)']
]

# Rename columns
df_gp_2020.columns = ['Date', 'Gold_Price']

# Remove commas from prices
df_gp_2020['Gold_Price'] = (
    df_gp_2020['Gold_Price']
    .astype(str)
    .str.replace(',', '')
)

# Convert Gold Price to numeric
df_gp_2020['Gold_Price'] = pd.to_numeric(
    df_gp_2020['Gold_Price'],
    errors='coerce'
)

# Convert Date column
df_gp_2020['Date'] = pd.to_datetime(
    df_gp_2020['Date'],
    format='%d-%b-%Y',
    errors='coerce'
)

# Remove invalid rows
df_gp_2020 = df_gp_2020[
    df_gp_2020['Date'].notna()
]

# Create Year-Month column
df_gp_2020['Date'] = df_gp_2020['Date'].dt.strftime('%Y-%m')

# Group by month and take average
df_gp_2020 = (
    df_gp_2020
    .groupby('Date', as_index=False)
    ['Gold_Price']
    .mean()
)

# Round values
df_gp_2020['Gold_Price'] = (
    df_gp_2020['Gold_Price']
    .round(2)
)

# Preview
df_gp_2020.head()
```

```python
df_bombay_gp_2000_2019 = pd.read_csv(
    "/content/Gold India Bombay Market Standard.csv",
    header=3 # Assuming the actual header is on the 4th line (index 3)
)
```

```python
df_bombay_gp_2000_2019
```

```python
# Rename first column
df_bombay_gp_2000_2019.columns = ['Raw_Data', 'Unused']

# Keep only first column
df_bombay_gp_2000_2019 = df_bombay_gp_2000_2019[['Raw_Data']]

# Remove rows with NaN
df_bombay_gp_2000_2019 = df_bombay_gp_2000_2019.dropna()

# Keep only rows containing dates
df_bombay_gp_2000_2019 = df_bombay_gp_2000_2019[
    df_bombay_gp_2000_2019['Raw_Data']
    .astype(str)
    .str.contains(r'\d{2}[-/]\d{2}[-/]\d{2,4}')
]

# Split date and price using tab separator
df_bombay_gp_2000_2019[
    ['Date', 'Gold_Price']
] = df_bombay_gp_2000_2019['Raw_Data'].str.split(
    '\t',
    expand=True
)

# Remove commas from price
df_bombay_gp_2000_2019['Gold_Price'] = (
    df_bombay_gp_2000_2019['Gold_Price']
    .astype(str)
    .str.replace(',', '')
)

# Convert Gold Price to numeric
df_bombay_gp_2000_2019['Gold_Price'] = pd.to_numeric(
    df_bombay_gp_2000_2019['Gold_Price'],
    errors='coerce'
)

# Convert Date column
df_bombay_gp_2000_2019['Date'] = pd.to_datetime(
    df_bombay_gp_2000_2019['Date'],
    errors='coerce'
)

# Remove invalid dates
df_bombay_gp_2000_2019 = df_bombay_gp_2000_2019[
    df_bombay_gp_2000_2019['Date'].notna()
]

# Keep only years 2000-2019
df_bombay_gp_2000_2019 = df_bombay_gp_2000_2019[
    (df_bombay_gp_2000_2019['Date'] >= '2000-01-01') &
    (df_bombay_gp_2000_2019['Date'] <= '2019-12-31')
]

# Convert to Year-Month
df_bombay_gp_2000_2019['Date'] = (
    df_bombay_gp_2000_2019['Date']
    .dt.strftime('%Y-%m')
)

# Monthly average gold price
df_bombay_gp_2000_2019 = (
    df_bombay_gp_2000_2019
    .groupby('Date', as_index=False)
    ['Gold_Price']
    .mean()
)

# Round values
df_bombay_gp_2000_2019['Gold_Price'] = (
    df_bombay_gp_2000_2019['Gold_Price']
    .round(2)
)

# Sort chronologically
df_bombay_gp_2000_2019 = (
    df_bombay_gp_2000_2019
    .sort_values('Date')
)

# Reset index
df_bombay_gp_2000_2019.reset_index(
    drop=True,
    inplace=True
)

# Preview
df_bombay_gp_2000_2019.head()
```
