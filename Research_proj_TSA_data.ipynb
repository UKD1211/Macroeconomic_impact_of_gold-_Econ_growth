```python
import pandas as pd
import numpy as np
```

```python
df_gold_reservs = pd.read_excel("/content/Gold Reserves USD mn Monthly India.xlsx")
```

```python
df_gold_reservs
```

```python
df_gold_reservs.columns = ["Date", "Gold_Reservs_USD_mn"]

```

```python
df_gold_reservs
```

```python
df_gold_reservs = df_gold_reservs.copy()

df_gold_reservs['Date'] = pd.to_datetime(
    df_gold_reservs['Date'],
    errors='coerce'
)

df_gold_reservs = df_gold_reservs[
    df_gold_reservs['Date'].notna()
]

df_gold_reservs['Gold_Reservs_USD_mn'] = pd.to_numeric(
    df_gold_reservs['Gold_Reservs_USD_mn'],
    errors='coerce'
)

df_gold_reservs = df_gold_reservs[
    df_gold_reservs['Date'] >= '2000-01-01'
]

df_gold_reservs.reset_index(drop=True, inplace=True)
```

```python
df_gold_volume = pd.read_excel("/content/International Liquidity Gold Volume.xlsx")
```

```python
df_gold_volume
```

```python
# Rename columns
df_gold_volume.columns = ['Date', 'Gold_Volume_mn_troy_oz']

# Convert Date column
df_gold_volume['Date'] = pd.to_datetime(
    df_gold_volume['Date'],
    errors='coerce'
)

# Keep only valid date rows
df_gold_volume = df_gold_volume[
    df_gold_volume['Date'].notna()
]

# Convert gold volume column to numeric
df_gold_volume['Gold_Volume_mn_troy_oz'] = pd.to_numeric(
    df_gold_volume['Gold_Volume_mn_troy_oz'],
    errors='coerce'
)

# Keep only data from 2000 onwards
df_gold_volume = df_gold_volume[
    df_gold_volume['Date'] >= '2000-01-01'
]

# Reset index
df_gold_volume.reset_index(drop=True, inplace=True)

# Preview cleaned dataframe
df_gold_volume.head()
```

```python
df_gold_volume
```

```python
df_gold_volume.to_excel(
    'clean_gold_volume.xlsx',
    index=False
)
```
