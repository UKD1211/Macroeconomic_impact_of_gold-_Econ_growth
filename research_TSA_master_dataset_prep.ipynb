```python
month_map = {
    'January':'01',
    'February':'02',
    'March':'03',
    'April':'04',
    'May':'05',
    'June':'06',
    'July':'07',
    'August':'08',
    'September':'09',
    'October':'10',
    'November':'11',
    'December':'12'
}

df_cpi_11_14_gold['Date'] = (
    df_cpi_11_14_gold['Year'].astype(str)
    + '-'
    + df_cpi_11_14_gold['Month'].map(month_map)
)

df_cpi_11_14_gold = df_cpi_11_14_gold[
    ['Date','Combined Index']
]

df_cpi_11_14_gold.columns = [
    'Date',
    'Gold_CPI'
]
```

```python
df_cpi_15_25_gold.columns = [
    'Year',
    'Month',
    'Item_Code',
    'Description',
    'Gold_CPI'
]

df_cpi_15_25_gold['Date'] = (
    df_cpi_15_25_gold['Year'].astype(str)
    + '-'
    + df_cpi_15_25_gold['Month'].map(month_map)
)

df_cpi_15_25_gold = df_cpi_15_25_gold[
    ['Date','Gold_CPI']
]
```

```python
df_gold_cpi = pd.concat(
    [
        df_cpi_11_14_gold,
        df_cpi_15_25_gold
    ],
    ignore_index=True
)

df_gold_cpi = (
    df_gold_cpi
    .sort_values('Date')
    .reset_index(drop=True)
)
```

```python
# Rename columns
df_inr_usd.columns = ['Date', 'USDINR']

# Convert Date column
df_inr_usd['Date'] = pd.to_datetime(
    df_inr_usd['Date'],
    errors='coerce'
)

# Convert exchange rate to numeric
df_inr_usd['USDINR'] = pd.to_numeric(
    df_inr_usd['USDINR'],
    errors='coerce'
)

# Remove invalid rows
df_inr_usd = df_inr_usd.dropna()

# Convert date to Year-Month
df_inr_usd['Date'] = (
    df_inr_usd['Date']
    .dt.strftime('%Y-%m')
)

# Monthly average
df_inr_usd = (
    df_inr_usd
    .groupby('Date', as_index=False)
    ['USDINR']
    .mean()
)

# Round values
df_inr_usd['USDINR'] = (
    df_inr_usd['USDINR']
    .round(2)
)

# Sort
df_inr_usd = (
    df_inr_usd
    .sort_values('Date')
    .reset_index(drop=True)
)

df_inr_usd.head()
```

```python
df_inr_usd
```

```python
df_master = pd.merge(
    df_gold_master,
    df_inr_usd,
    on='Date',
    how='left'
)
```

```python
df_master
```

```python
# Rename columns
df_repo_rate.columns = ['Date', 'Repo_Rate']

# Convert Date
df_repo_rate['Date'] = pd.to_datetime(
    df_repo_rate['Date'],
    errors='coerce'
)

# Keep only valid date rows
df_repo_rate = df_repo_rate[
    df_repo_rate['Date'].notna()
]

# Convert repo rate to numeric
df_repo_rate['Repo_Rate'] = pd.to_numeric(
    df_repo_rate['Repo_Rate'],
    errors='coerce'
)

# Keep from 2000 onward
df_repo_rate = df_repo_rate[
    df_repo_rate['Date'] >= '2000-01-01'
]

# Convert to YYYY-MM
df_repo_rate['Date'] = (
    df_repo_rate['Date']
    .dt.strftime('%Y-%m')
)

# Sort
df_repo_rate = (
    df_repo_rate
    .sort_values('Date')
    .reset_index(drop=True)
)

df_repo_rate.head()
```

```python
# Convert MultiIndex to dataframe
df_cpi_overall_master = pd.DataFrame(
    df_cpi_overall_13_25.index.tolist()
)

# Use first row as column names
df_cpi_overall_master.columns = df_cpi_overall_master.iloc[0]

# Remove header row
df_cpi_overall_master = df_cpi_overall_master.iloc[1:]

# Reset index
df_cpi_overall_master = df_cpi_overall_master.reset_index(drop=True)

df_cpi_overall_master.head()
```

```python
month_cols = [
    'Jan','Feb','Mar','Apr','May','Jun',
    'Jul','Aug','Sep','Oct','Nov','Dec'
]

df_cpi_overall_master = df_cpi_overall_master.melt(
    id_vars='Year',
    value_vars=month_cols,
    var_name='Month',
    value_name='Overall_CPI'
)
```

```python
month_map = {
    'Jan':'01',
    'Feb':'02',
    'Mar':'03',
    'Apr':'04',
    'May':'05',
    'Jun':'06',
    'Jul':'07',
    'Aug':'08',
    'Sep':'09',
    'Oct':'10',
    'Nov':'11',
    'Dec':'12'
}

df_cpi_overall_master['Date'] = (
    df_cpi_overall_master['Year'].astype(str)
    + '-'
    + df_cpi_overall_master['Month'].map(month_map)
)

df_cpi_overall_master = (
    df_cpi_overall_master[['Date','Overall_CPI']]
    .sort_values('Date')
    .reset_index(drop=True)
)
```

```python
import numpy as np
import pandas as pd

# Convert Overall_CPI to numeric just in case
df_master['Overall_CPI'] = pd.to_numeric(
    df_master['Overall_CPI'],
    errors='coerce'
)

# Year-on-Year Inflation (%)
df_master['Inflation_YoY'] = (
    df_master['Overall_CPI']
    .pct_change(periods=12)
    * 100
)

# Month-on-Month Inflation (%)
df_master['Inflation_MoM'] = (
    df_master['Overall_CPI']
    .pct_change(periods=1)
    * 100
)

# Round values
df_master['Inflation_YoY'] = (
    df_master['Inflation_YoY']
    .round(2)
)

df_master['Inflation_MoM'] = (
    df_master['Inflation_MoM']
    .round(2)
)

# Optional: replace NaN with '---' for display/reporting
df_master['Inflation_YoY'] = (
    df_master['Inflation_YoY']
    .fillna('---')
)

df_master['Inflation_MoM'] = (
    df_master['Inflation_MoM']
    .fillna('---')
)

df_master.head()
```
