```python
import pandas as pd

all_fpi = []

for year in range(2002, 2027):

    file = f"{year}.xls"

    # Read table
    df = pd.read_html(file)[0]

    # Keep first 12 monthly rows
    temp = df.iloc[:12].copy()

    # Rename columns (adjust if needed)
    temp = temp.iloc[:, [0, 3]]

    temp.columns = [
        "Month",
        "FPI_Net_INR_Crore"
    ]

    # Create Date column
    month_map = {
        'January':'01',
        'February':'02',
        'March':'03',
        'April':'04',
        'May':'05',
        'June':'06',
        'July':'07',
        'August':'08',
        'September':'09',
        'October':'10',
        'November':'11',
        'December':'12'
    }

    temp["Date"] = (
        str(year)
        + "-"
        + temp["Month"].map(month_map)
    )

    temp = temp[
        ["Date", "FPI_Net_INR_Crore"]
    ]

    all_fpi.append(temp)

# Combine all years
df_fpi_master = pd.concat(
    all_fpi,
    ignore_index=True
)

print(df_fpi_master.head())
print(df_fpi_master.tail())
```

```python
df_fpi_master.to_excel("FPI_Master.xlsx", index=False)
```

```python
import pandas as pd

# The original code produced a SyntaxError because '01' was treated as a numeric literal with a leading zero,
# which is not allowed in Python for decimal integers. '2017-01' was interpreted as 2017 minus the integer 01 (which is 1),
# resulting in 2016. To compare against a date string, it needs to be explicitly a string like '2017-01'.
#
# However, for accurate chronological filtering and to handle potential non-date entries (like NaN or text)
# in the 'Date' column of `df_fpi_master`, it's best to clean the data and convert the 'Date' column to datetime objects.

# Step 1: Clean df_fpi_master
# Create a copy to avoid modifying the original df_fpi_master directly in case it's needed later in its original form.
df_fpi_master_cleaned = df_fpi_master.copy()

# Convert 'FPI_Net_INR_Crore' to numeric, coercing errors to NaN, then drop rows with NaN in this column.
# This handles any descriptive text or invalid numbers in this column.
df_fpi_master_cleaned['FPI_Net_INR_Crore'] = pd.to_numeric(df_fpi_master_cleaned['FPI_Net_INR_Crore'], errors='coerce')
df_fpi_master_cleaned = df_fpi_master_cleaned.dropna(subset=['FPI_Net_INR_Crore']).copy()

# Convert 'Date' column to datetime, coercing errors to NaT (Not a Time), then drop rows with NaT.
# This handles any NaN or descriptive text in the 'Date' column.
df_fpi_master_cleaned['Date'] = pd.to_datetime(df_fpi_master_cleaned['Date'], format='%Y-%m', errors='coerce')
df_fpi_master_cleaned = df_fpi_master_cleaned.dropna(subset=['Date']).copy()

# Step 2: Define the start date for filtering as a datetime object
start_date = pd.to_datetime('2017-01')

# Step 3: Filter the DataFrame chronologically
filtered_df = df_fpi_master_cleaned[df_fpi_master_cleaned['Date'] >= start_date]

# Display the filtered DataFrame
filtered_df
```

```python
df_fpi_master_cleaned = df_fpi_master_cleaned[~(df_fpi_master_cleaned['Date'] >= start_date)]
```
