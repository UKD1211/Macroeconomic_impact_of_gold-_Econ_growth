```python
usdinr_ci = forecast_USDINR_df.rename(columns={
    "Unnamed: 0": "Date",
    "lower USDINR": "USDINR_Lower",
    "upper USDINR": "USDINR_Upper"
})[["Date","USDINR_Lower","USDINR_Upper"]]
```

```python
repo_ci = forecast_Repo_Rate_df.rename(columns={
    "Unnamed: 0":"Date",
    "Lower Repo Rate":"Repo_Rate_Lower",
    "Upper Repo Rate":"Repo_Rate_Upper"
})[["Date","Repo_Rate_Lower","Repo_Rate_Upper"]]
```

```python
gold_ci = forecast_GoldRes_df.rename(columns={
    "Unnamed: 0":"Date",
    "Lower":"GoldReserve_Lower",
    "Upper":"GoldReserve_Upper"
})[["Date","GoldReserve_Lower","GoldReserve_Upper"]]
```

```python
gdp_ci = forecast_GDP_df.rename(columns={
    "Unnamed: 0":"Date",
    "Lower GDP":"GDP_Lower",
    "Upper GDP":"GDP_Upper"
})[["Date","GDP_Lower","GDP_Upper"]]
```

```python
inflation_ci = forecast_inflation_df.rename(columns={
    "Inflation_Lower_CI":"Inflation_Lower",
    "Inflation_Upper_CI":"Inflation_Upper"
})[["Date","Inflation_Lower","Inflation_Upper"]]
```

```python
fpi_ci = forecast_FPI_df.rename(columns={
    "Lower_CI":"FPI_Lower",
    "Upper_CI":"FPI_Upper"
})[["Date","FPI_Lower","FPI_Upper"]]
```

```python
dfs = [usdinr_ci, repo_ci, gold_ci, gdp_ci, inflation_ci, fpi_ci]

for df in dfs:
    df["Date"] = pd.to_datetime(df["Date"])
```

```python
end_date = pd.Timestamp("2027-12-01")

dfs = [
    df[df["Date"] <= end_date].copy()
    for df in dfs
]
```

```python
ci_master = dfs[0]

for df in dfs[1:]:
    ci_master = ci_master.merge(
        df,
        on="Date",
        how="outer"
    )

ci_master = ci_master.sort_values("Date").reset_index(drop=True)
```
